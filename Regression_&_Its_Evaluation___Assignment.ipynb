Question 1: What is Simple Linear Regression?    
Answer- Simple linear regression is a statistical method used to model the relationship between two continuous variables. It aims to find a linear equation, represented as a straight line, that best fits the data and can be used to predict the value of a dependent variable based on the value of a single independent variable.   

Question 2: What are the key assumptions of Simple Linear Regression?  
Answer- Simple linear regression, to be a valid and reliable model, relies on several key assumptions about the data and the error term. Violating these assumptions can lead to misleading or inaccurate results.

1. Linearity
There must be a linear relationship between the independent variable (X) and the dependent variable (Y). The model assumes that the line of best fit is a straight line, so if the true relationship is non-linear (e.g., curved), the model will not accurately represent the data.

2. Independence of Errors
The errors (residuals) of the model must be independent of each other. This means that the error for one observation should not be correlated with the error for any other observation. This assumption is particularly important in time-series data, where consecutive errors can be dependent on one another (a condition called autocorrelation).

3. Homoscedasticity
This assumption, also known as constant variance, states that the variance of the errors is constant across all levels of the independent variable. In simple terms, the spread of the residuals should be uniform. If the variance of the residuals increases or decreases as the value of the independent variable changes (a condition called heteroscedasticity), it can affect the reliability of the model's predictions.

4. Normality of Errors
The errors (residuals) should be normally distributed. This assumption is crucial for making statistical inferences, such as hypothesis testing and constructing confidence intervals. While minor deviations from normality are often acceptable, significant skewness or outliers can impact the validity of these statistical tests.    

Question 3: What is heteroscedasticity, and why is it important to address in regression
models?  
Answer: Heteroscedasticity is a violation of a key assumption in linear regression where the variance of the residuals (errors) is not constant across all levels of the independent variable(s). It literally means "unequal scatter" (from Greek "hetero" meaning different and "skedasis" meaning scattering). This results in a funnel or fan shape in a residual plot, where the spread of the data points around the regression line increases or decreases as the independent variable changes.



Why It's a Problem
While heteroscedasticity doesn't bias the regression coefficients themselves (the estimated slope and intercept are still, on average, correct), it makes them inefficient and their standard errors unreliable. This is a major issue because standard errors are used to perform statistical tests and build confidence intervals, which are essential for drawing conclusions from your model.


Here are the specific consequences:

Inaccurate Standard Errors: The standard errors of the coefficients are likely to be underestimated. This means your confidence intervals will be too narrow, and you'll be more likely to find a statistically significant relationship that doesn't actually exist in the population (Type I error).

Unreliable P-values and Hypothesis Tests: Because the standard errors are incorrect, the t-tests and F-tests used to assess the significance of the coefficients and the overall model are no longer valid. You might conclude that a variable is a significant predictor when it isn't.


Inefficient OLS Estimators: Ordinary Least Squares (OLS) regression is no longer the "Best Linear Unbiased Estimator" (BLUE). This means there are other methods that could produce more precise and reliable coefficient estimates.

How to Address It
Detecting heteroscedasticity is often done by plotting the residuals against the fitted values or the independent variable(s) and looking for a fan shape. Formal tests like the Breusch-Pagan test or the White test can also be used.


Once detected, you can address heteroscedasticity by:

Using Robust Standard Errors: This is the most common and often easiest solution. It corrects the standard errors to be valid even in the presence of heteroscedasticity, without changing the regression coefficients.

Transforming Variables: Applying a mathematical transformation, such as a logarithmic or square root transformation, to the dependent variable can sometimes stabilize the variance and eliminate the fan shape.

Using Weighted Least Squares (WLS): This method assigns a weight to each data point, giving less importance to observations with higher error variance and more importance to those with lower variance.

Question 6: Implement a Python program to fit a Simple Linear Regression model to
the following sample data:
● X = [1, 2, 3, 4, 5]
● Y = [2.1, 4.3, 6.1, 7.9, 10.2]
Plot the regression line over the data points.   
Answer-

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# The sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.1, 4.3, 6.1, 7.9, 10.2])

# Create and fit the model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Get the coefficients
intercept = model.intercept_
slope = model.coef_[0]

print(f"Intercept (β₀): {intercept:.4f}")
print(f"Slope (β₁): {slope:.4f}")

# Plot the data and the regression line
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='Regression line')
plt.title('Simple Linear Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

Question 7: Fit a Multiple Linear Regression model on this sample data:
● Area = [1200, 1500, 1800, 2000]
● Rooms = [2, 3, 3, 4]
● Price = [250000, 300000, 320000, 370000]   
Check for multicollinearity using VIF and report the results.  
Answer-


```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
from statsmodels.stats.outliers_influence import variance_inflation_factor
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Data
data = {
    'Area': [1200, 1500, 1800, 2000],
    'Rooms': [2, 3, 3, 4],
    'Price': [250000, 300000, 320000, 370000]
}
df = pd.DataFrame(data)

# Define independent and dependent variables
X = df[['Area', 'Rooms']]
y = df['Price']

# Add a constant to the independent variables for the intercept
X = sm.add_constant(X)

# Fit the Multiple Linear Regression model
model = sm.OLS(y, X).fit()

# Print the model summary
print("Multiple Linear Regression Model Summary:\n")
print(model.summary())
print("\n" + "="*50 + "\n")

# Check for multicollinearity using VIF
print("Variance Inflation Factor (VIF) Results:")
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(len(X.columns))]
print(vif_data)

# 3D Plot of the regression plane
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(df['Area'], df['Rooms'], df['Price'], color='blue', label='Data Points')

# Create a meshgrid for the regression plane
x_surf = np.arange(df['Area'].min(), df['Area'].max(), 50)
y_surf = np.arange(df['Rooms'].min(), df['Rooms'].max(), 0.1)
x_surf, y_surf = np.meshgrid(x_surf, y_surf)

# Calculate the predicted price for each point on the meshgrid
price_pred_surf = model.params['const'] + model.params['Area'] * x_surf + model.params['Rooms'] * y_surf

# Plot the regression plane
ax.plot_surface(x_surf, y_surf, price_pred_surf, color='red', alpha=0.3, label='Regression Plane')

# Labels and title
ax.set_xlabel('Area (sq ft)')
ax.set_ylabel('Rooms')
ax.set_zlabel('Price')
ax.set_title('Multiple Linear Regression: Price vs. Area and Rooms')

# Custom legend for the 3D plot
from matplotlib.lines import Line2D
legend_elements = [Line2D([0], [0], marker='o', color='w', label='Data Points',
                          markerfacecolor='blue', markersize=10),
                   Line2D([0], [0], linestyle='-', color='red', lw=4, label='Regression Plane')]
ax.legend(handles=legend_elements)

plt.show()
```

Question 8: Implement polynomial regression on the following data:
● X = [1, 2, 3, 4, 5]
3  
● Y = [2.2, 4.8, 7.5, 11.2, 14.7]  
Fit a 2nd-degree polynomial and plot the resulting curve.  

Answer-

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

# The sample data
X = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)
Y = np.array([2.2, 4.8, 7.5, 11.2, 14.7])

# Create polynomial features up to degree 2
poly_features = PolynomialFeatures(degree=2, include_bias=False)
X_poly = poly_features.fit_transform(X)

# Create and fit the linear regression model on the polynomial features
model = LinearRegression()
model.fit(X_poly, Y)

# Predict Y values
Y_pred = model.predict(X_poly)

# Get the coefficients
intercept = model.intercept_
coef = model.coef_

print(f"Intercept (β₀): {intercept:.4f}")
print(f"Coefficients (β₁ and β₂): {coef}")
print(f"Polynomial equation: Y = {intercept:.2f} + {coef[0]:.2f}X + {coef[1]:.2f}X^2")

# Plot the original data and the polynomial curve
plt.scatter(X, Y, color='blue', label='Data points')
plt.plot(X, Y_pred, color='red', label='2nd-degree Polynomial Fit')
plt.title('Polynomial Regression')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend()
plt.grid(True)
plt.show()
```

Question 9: Create a residuals plot for a regression model trained on this data:
● X = [10, 20, 30, 40, 50]  
● Y = [15, 35, 40, 50, 65]  
Assess heteroscedasticity by examining the spread of residuals    
Answer-

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# The sample data
X = np.array([10, 20, 30, 40, 50]).reshape(-1, 1)
Y = np.array([15, 35, 40, 50, 65])

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, Y)

# Predict Y values
Y_pred = model.predict(X)

# Calculate the residuals
residuals = Y - Y_pred

# Create the residuals plot
plt.figure(figsize=(8, 6))
plt.scatter(X, residuals, color='blue', label='Residuals')
plt.axhline(y=0, color='red', linestyle='--', label='Zero Residuals Line')
plt.title('Residuals Plot')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Residuals')
plt.legend()
plt.grid(True)
plt.show()

print("Residuals:")
print(residuals)
```

Question 10: Imagine you are a data scientist working for a real estate company. You
need to predict house prices using features like area, number of rooms, and location.
However, you detect heteroscedasticity and multicollinearity in your regression
model. Explain the steps you would take to address these issues and ensure a robust
model.   
Answer- As a data scientist, encountering heteroscedasticity and multicollinearity is a common challenge. Here are the steps I would take to address each issue and build a more robust predictive model for house prices.

Addressing Heteroscedasticity
Heteroscedasticity means the variance of the errors is not constant. While it doesn't bias the coefficient estimates, it makes their standard errors unreliable, which can lead to incorrect conclusions about which features are statistically significant.

Detection: First, I would confirm the presence of heteroscedasticity by creating a residuals plot. I would plot the residuals against the predicted values. A "fan" or "cone" shape in the plot would indicate heteroscedasticity. I would also run a formal statistical test, like the Breusch-Pagan test or White's test, for a quantitative confirmation.

Using Robust Standard Errors: The most straightforward and effective solution is to use robust standard errors (also known as Huber-White standard errors). This method adjusts the standard errors to be valid even when the heteroscedasticity is present. It doesn't change the coefficient estimates but provides a more reliable basis for p-values and confidence intervals. This is my preferred approach as it's simple to implement and doesn't require transforming the data.

Variable Transformation: If robust standard errors aren't sufficient, I would consider transforming the dependent variable. For skewed data like house prices, a logarithmic transformation (log(Price)) is often effective. This can help stabilize the variance and make the relationship more linear. However, the downside is that the model's predictions are now in logarithmic units and must be converted back to the original scale, which can complicate interpretation.

Addressing Multicollinearity
Multicollinearity occurs when two or more independent variables in the model are highly correlated. This makes it difficult to isolate the unique effect of each predictor on the dependent variable, leading to unstable and hard-to-interpret coefficient estimates.

Detection: I would start by calculating the Variance Inflation Factor (VIF) for each predictor. A VIF value greater than 5 or 10 is a strong indicator of multicollinearity. I would also examine a correlation matrix of the independent variables to see which pairs are highly correlated.

Feature Selection: The most direct way to solve multicollinearity is to remove one of the correlated variables. For example, if "number of bedrooms" and "living area" are highly correlated (which they often are), I would consider keeping only the variable that is a better predictor or is more interpretable.

Principal Component Analysis (PCA): For more complex cases with multiple correlated variables, I would use Principal Component Analysis (PCA). PCA transforms the original set of correlated features into a new, smaller set of uncorrelated features called principal components. These components can then be used in the regression model, eliminating multicollinearity by design. The trade-off is that the principal components are less interpretable than the original features.

Regularization Techniques: I would also consider using regularization methods like Ridge Regression or Lasso Regression. Ridge regression is specifically designed to handle multicollinearity by adding a penalty term that shrinks the coefficients of correlated variables, reducing their variance. Lasso regression can do the same and also performs feature selection by shrinking some coefficients to zero.
